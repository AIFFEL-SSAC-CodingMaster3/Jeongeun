# 원형 데크 디자인

요구사항

1. 생성자(크기 지정 : k)
---
2. 데크 처음 아이템 추가, 성공하면 true 리턴
3. 데크 처음에 아이템 삭제, 성공하면 true 리턴
4. 데크 처음에 아이템 가져오기, 비어있으면 -1 리턴
---
5. 데크 마지막 아이템 추가, 성공하면 true 리턴
6. 데크 마지막 아이템 삭제, 성공하면 true 리턴
7. 데크 마지막 아이템 가져오기, 비어있다면 -1 리턴
---
8. 4와 7을 수행할 때, 데크가 비었는지 여부를 알 수 있도록 판별하여야 함
9. 데크가 '가득 찼는지' 1의 생성자 크기 k를 가져와 확인함

```python
class MyCircularDeque:

    def __init__(self, k: int):
        self.deque = [None]*k
        self.maxlen = k
        self.size = 0
        
        #앞, 뒤 판별
        self.head=0
        self.tail=0

    def insertFront(self, value: int) -> bool:
        
        if self.isEmpty():
            self.deque[self.head]=value
            self.size += 1
        
        #deque에 들어갈 자리가 남아있는지 먼저 체크
        elif not self.isFull():
            #전부 차 있으면 왼쪽 끝으로 들어가야 함
            self.head = (self.head-1) % self.maxlen
            
            self.deque[self.head]=value
            self.size += 1
            return True
        
        else:
            return False

    def insertLast(self, value: int) -> bool:
       
        if self.isEmpty():
            self.deque[self.tail] = value
            self.size += 1
            return True
        
        elif not self.isFull():
            self.tail = (self.tail+1)%self.maxlen
            self.deque[self.tail] = value
            self.size += 1
            
            return True
        
        else:
            return False

    def deleteFront(self) -> bool:
        
        #비어있으면 출력할 것도 없다!
        if not self.isEmpty():
            
            #값을 None으로 변경(삭제)
            self.deque[self.head] = None
            
            #사이즈가 1이 아니라면 헤드를 옮겨줘야 함
            if self.size != 1 :
                self.head = (self.head +1)% self.maxlen
                
                self.size -=1
                return True
        else:
            return False
        

    def deleteLast(self) -> bool:
       
        if not self.isEmpty():
            self.deque[self.tail] = None
            
            if self.size != 1:
                self.tail = (self.tail -1) % self.maxlen
                
                self.size -=1
                return True
        else:
            return False
        

    def getFront(self) -> int:
        return self.deque[self.head] if not self.isEmpty() else -1
        

    def getRear(self) -> int:
        return self.deque[self.tail] if not self.isEmpty() else -1

    def isEmpty(self) -> bool:
        return self.size == 0

    def isFull(self) -> bool:
        return self.size == self.maxlen
```
